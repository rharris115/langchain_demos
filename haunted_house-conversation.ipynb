## A Haunted House

```python
class Ghost(BaseModel):
    name_in_afterlife: str
    description_in_afterlife: str
    name_when_living: str
    description_when_living: str
    type_of_ghost: str
    scariness: int
    sex: str
    year_of_birth: int
    year_of_demise: int
    description_of_demise: str
    conditions_for_moving_on: str
    preferred_victims_to_scare: str
    preferred_method_to_scare: str


class HauntedHouse(BaseModel):
    name: str
    description: str
    type_of_house: str
    location: str
    year_built: int
    last_year_inhabited_if_abandoned: int
    ghosts: list[Ghost]


pydantic_parser = PydanticOutputParser(pydantic_object=HauntedHouse)
```

```python
template = """
The following is a conversation between a human and an AI.
The AI provides a haunted house with lots of specific details from its context.

Each ghost is rated on scariness according to the following definition.

{scariness_definition}

{format_instructions}

The AI follows this format strictly, lists each ghost individually, and makes no other comment before or after.

Current conversation:
{history}
Human: {input}
AI Assistant:"""


prompt = PromptTemplate(
    template=template,
    input_variables=["history", "input"],
    partial_variables={
        "format_instructions": pydantic_parser.get_format_instructions(),
        "scariness_definition": str(scariness_definition),
    },
)

conversation = ConversationChain(
    prompt=prompt,
    llm=llm,
    verbose=True,
    memory=ConversationBufferWindowMemory(k=1, ai_prefix="AI Assistant"),
)
```

```python
conversation.run(
    "Can you give me a haunted house in Covington Georgia with a few ghosts."
)
JSON(json.loads(conversation.memory.buffer_as_messages[-1].content))
```

```python
conversation.run("Can you add six more ghosts?")
JSON(json.loads(conversation.memory.buffer_as_messages[-1].content))
```

```python
conversation.run("Can you add yet another three more ghosts?")
JSON(json.loads(conversation.memory.buffer_as_messages[-1].content))
```

### Comments: Fewer tokens for more ghosts!

While this is fun, it's chat message histories really just accumulate previous histories into a single prompt. After a time, this will make it difficult to keep the token count under the total limit for input plus output. In our case, we couldn't force `ChatGPT` to provide additional ghosts because there just weren't enough tokens left.

There are much more sophisticated ways to provide context to `ChatGPT` to economise on tokens, but that is topic for another discussion.
